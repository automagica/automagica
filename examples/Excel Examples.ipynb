# Automating Excel with Automagica
MicrosoftÂ® Office Excel can be easily automated with Automagica. We will show you how in this notebook.

```python
from automagica.activities import ExcelApplication
```

Let's first open up Excel:

```python
excel = ExcelApplication()
```

Alternatively, you can run Excel in the background with:

```python
excel = ExcelApplication(visible=False)
```

But for the purpose of this example, we're going to run it with the default value for `visible`.

Excel is now open:


```python
workbook = excel.new_workbook()
```

We just created our first workbook!


Let's add a sheet to this workbook:

```python
worksheet = excel.add_worksheet(workbook, name='My Sheet')
```

We now have a new worksheet called `My Sheet`! 


Let's put some data in there

```python
excel.write_range(worksheet, 'A1:B2', 'data')
```

Looks like there's a lot of data in this Excel now :-). Let's get a bit more creative.

```python
my_data = (
    ('Column 1', 'Column 2', 'Column 3'),
    (1, 2, 3),
    (4, 5, 6)
)
excel.write_range(worksheet, 'A1:C3', my_data)
```

Not very creative, maybe let's try writing cells directly:

```python
for i in range(1, 4):
    for j in range(1, 4):
        excel.write_cell(worksheet, i, j, i*j)
```
